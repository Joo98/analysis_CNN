```python
import pandas as pd                # 데이터를 저장하고 처리하는 패키지
import matplotlib as mpl           # 그래프를 그리는 패키지
import matplotlib.pyplot as plt    # 그래프를 그리는 패키지
```

```python
Relu_50 = pd.read_csv('result/50_Relu.csv')
Tanh_50 = pd.read_csv('result/50_Tanh.csv')
```

```python
x_Relu_50 = list(Relu_50['Step'])
x_Tanh_50 = list(Tanh_50['Step'])
```

```python
y_Relu_50 = list(Relu_50['Value'])
y_Tanh_50 = list(Tanh_50['Value'])
```

```python
plt.plot(x_Relu_50,y_Relu_50,label ='ResNet50 with Relu')
plt.plot(x_Relu_50,y_Tanh_50,label ='ResNet50 with Tanh')
plt.legend()
plt.savefig('result/Resnet50_Activation.png')
```

```python
Relu_152 = pd.read_csv('result/152_Relu.csv')
Tanh_152 = pd.read_csv('result/152_Tanh.csv')
x_Relu_152 = list(Relu_152['Step'])
x_Tanh_152 = list(Tanh_152['Step'])
y_Relu_152 = list(Relu_152['Value'])
y_Tanh_152 = list(Tanh_152['Value'])
plt.plot(x_Relu_152,y_Relu_152,label ='ResNet152 with Relu')
plt.plot(x_Relu_152,y_Tanh_152,label ='ResNet152 with Tanh')
plt.legend()
plt.savefig('result/Resnet152_Activation.png')
```

```python
Relu_152 = pd.read_csv('result/152_Relu.csv')
without_Residual_152 = pd.read_csv('result/152_Without_Relu_score.csv')
x_Relu_152 = list(Relu_152['Step'])
x_without_Residual_152 = list(without_Residual_152['Step'])
y_Relu_152 = list(Relu_152['Value'])
y_without_Residual_152 = list(without_Residual_152['Value'])
plt.plot(x_Relu_152,y_Relu_152,label ='ResNet152 with Residual')
plt.plot(x_without_Residual_152,y_without_Residual_152,label ='ResNet152 without Residual')
plt.legend()
plt.savefig('result/Resnet152_Residual_score.png')
```
